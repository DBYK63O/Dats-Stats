Visualizing Loans Awarded by Kiva
In this project you'll visualize insights using a dataset from Kaggle. The dataset contains information about loans awarded by the non-profit Kiva.Visualizing Loans Awarded by Kiva.

Step 1: Import Necessary Python Modules

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

Step 2: Ingest The Data

```python
df = pd.read_csv('kiva_data.csv')
print(df.head())
```

Step 3: Examine The Data

```python
print(df.head(25))
```

Step 4: Bar Charts

```python
import matplotlib.ticker as mtick

f, ax = plt.subplots(figsize=(15, 10))
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick) 

sns.barplot(data=df, x="country", y = "loan_amount")
```

Step 5: Learn More By Using hue In Your Visualization

```python
f, ax = plt.subplots(figsize=(15, 10))
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

sns.barplot(data=df, x="country", y="loan_amount", hue = 'gender')
```

Step 6: Styling

```python
# Set color palette
sns.set_palette("Pastel1")

# Set style
sns.set_style('whitegrid')

# Create figure and axes (no need to use the previous syntax, as the y-label ticks aren't going to be formatted)
plt.figure(figsize=(25, 15))

# Add a title
ax.set_title("Average Loan Amount by Gender per Country")

# Use Seaborn to create the bar plot
f, ax = plt.subplots(figsize=(25, 15))
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
sns.barplot(data=df, x="country", y="loan_amount", hue="gender")
plt.show()
```

Step 7: Box Plots With Kiva Data

```python
plt.figure(figsize=(16, 10))
sns.set_palette("colorblind")
sns.boxplot(data=df, x="country", y="loan_amount", hue = 'gender')
plt.show()
```

    # Cambodia has the highest mean and fullest range near the top


Step 8: Box Plot by Activity

```python
plt.figure(figsize=(16, 10))
sns.set_palette("colorblind")
sns.boxplot(data=df, x="activity", y="loan_amount")
plt.show()
```

Reflection Questions: 
    What does this visualization reveal that previous ones did not?
    # It shows you the loan amounts by activity, rather than country. 
    # On average, "Farming" receives the largest loan amount and also has the widest distribution.
    # While General Store activity has the lowest loan amount and narrowest distribution, outliers outstanding.
    # This may reflect the countries current stage of economic development, being agriculturally based and labor intesnsive.

Step 9: Violin Plots

```python
plt.figure(figsize=(16, 10))
sns.set_palette("bright")
sns.violinplot(data=df, x="activity", y="loan_amount")
```

Create a violin plot that visualizes the distribution of loan amount by country.

```python
plt.figure(figsize=(16, 10))
sns.set_palette("bright")
sns.violinplot(data=df, x="country", y="loan_amount")
```

Step 10: Split Violin Plots

```python
plt.figure(figsize=(18, 12))

sns.violinplot(data=df, x="country", y = "loan_amount", hue="gender")
plt.show()
```

```python
plt.figure(figsize=(18, 12))

sns.violinplot(data=df, x="country", y = "loan_amount", hue="gender", split=True)
plt.show()
```
